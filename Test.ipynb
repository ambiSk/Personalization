```python
from selfie import ImageAPI
import time
from glob import glob
api = ImageAPI()
import matplotlib.pyplot as plt
from PIL import Image
```

```python
for jpg in glob("..\\*.jpg"):
    arr = Image.open(jpg)
    
    plt.imshow(arr)
    plt.title(jpg)
    plt.show()
    if "..\\g" in jpg:
        print('female')
        sessionID = api.send_request(jpg, open(jpg, 'rb'),'female')
        time.sleep(3)
        value = api.get_data(sessionID)
        print(value['FemaleEyes_Ball_R']['color'], value['FemaleEyes_Ball_L']['color'])
    else:
        print('male')
        sessionID = api.send_request(jpg, open(jpg, 'rb'),'male')
        time.sleep(3)
        value = api.get_data(sessionID)
        print(value['MaleEyes_Ball_R']['color'], value['MaleEyes_Ball_L']['color'])
```
